# Cross match LEGACY and Lyttle 2021

```python
import os
import pandas as pd

from astropy.io import ascii
```

```python
data_dir = 'data'

legacy_table = ascii.read(os.path.join(data_dir, 'LEGACY/globals.dat'))
lyttle_table = ascii.read(os.path.join(data_dir, 'Lyttle21/table1.txt'))
```

```python
legacy_df = legacy_table.to_pandas()
lyttle_df = lyttle_table.to_pandas()
```

```python
legacy_df.head()
```

```python
lyttle_df.head()
```

```python
lyttle_df['KIC'] = lyttle_df['name'].str.strip('KIC')
lyttle_df.head()
```

```python
df = pd.merge(lyttle_df, legacy_df, on='KIC', suffixes=('_L21', '_LEG'))
df
```

```python
df['KIC'].to_csv(os.path.join(data_dir, 'kic_ids.csv'), index=False)
```
